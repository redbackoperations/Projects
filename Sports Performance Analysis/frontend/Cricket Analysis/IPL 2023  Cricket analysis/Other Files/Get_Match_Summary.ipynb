```python
#Importing the library
import pandas as pd
```

```python
#Link for IPL 2023 match summaries(ESPN Cricinfo website)
url = 'https://www.espncricinfo.com/records/season/team-match-results/2023-2023?trophy=117'
```

```python
#Reading the html file(website)
all_tables = pd.read_html(url)
```

```python
#Cheking the type of the variable
type(all_tables)
```

```python
#Exporting the data into csv file
df = all_tables[0]

df.to_csv('Espncricinfo.csv')
```
